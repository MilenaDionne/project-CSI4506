### Formating des datasets

```python
def formatDataSet (dataset): 
    predict_1week = []
    future1week = []
    day1before = []
    day2before = []
    day3before = []
    day4before = []
    day5before = []
    day6before = []
    day7before = []
    count = 7
    while(count < len(dataset)-14):
        future1week.append(dataset[' Close/Last'][count])
        day1before.append(dataset[' Close/Last'][count+7])
        day2before.append(dataset[' Close/Last'][count+8])
        day3before.append(dataset[' Close/Last'][count+9])
        day4before.append(dataset[' Close/Last'][count+10])
        day5before.append(dataset[' Close/Last'][count+11])
        day6before.append(dataset[' Close/Last'][count+12])
        day7before.append(dataset[' Close/Last'][count+13])
        count+= 1
    data = {'future1week':future1week, 
            'day1before':day1before,
            'day2before':day2before,
            'day3before':day3before,
            'day4before':day4before,
            'day5before':day5before,
            'day6before':day6before,
            'day7before':day7before}
    return pd.DataFrame(data)
```

```python
type(aapl)
new_format_aapl = formatDataSet(aapl)

```
